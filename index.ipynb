```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def dydt(t, y):
    x, v = y
    return [v, -x]

y0 = [1, 0]

t_eval = np.linspace(0, 10, 300)

sol = solve_ivp(dydt, (0, 10), y0, t_eval=t_eval)

plt.plot(sol.t, sol.y[0])
plt.show()
```
