## Laplace Equation on a Disk of Radius 1

We solved the Laplace equation on a disk of radius $1$: $r<1$, with boundary conditions

$$u(1,\theta) = \cos(2\theta)$$

The solution is

$$ u(r, \theta) = r^2\cos(2\theta) $$

We can plot this as a colour map in polar coordinates:

```python
import numpy as np
from matplotlib import pyplot as plt


# set up a grid in r-theta coordinates
zeniths = np.linspace(0, 1, 20) # r values
azimuths = np.linspace(0, 2*np.pi, 50) # theta values
r, theta = np.meshgrid(zeniths, azimuths) # turns 1D arrays into 2D grids

u = r**2 * np.cos(2*theta) # function

fig, ax = plt.subplots(subplot_kw = dict(projection = 'polar')) # polar plot
ax.contourf(theta, r, u, 30, cmap = 'viridis') # plot u with 30 contour levels (smoothness/detail), 'viridis' colour map

plt.show()
```

## Laplace Equation on an Annulus

We also solved the Laplace equation with the following boundary conditions:

$$1~ < ~r~ < ~2$$  $$u(1,\theta)=0$$ $$u(2,\theta)=\cos(2\theta)$$

The solution is

$$ u(r,\theta) = \frac{4}{15}\left(r^2 - \frac{1}{r^2}\right)\cos(2\theta) $$

The following plots are more customized than the previous one:

```python
zeniths = np.linspace(1, 2, 20)
azimuths = np.linspace(0, 2*np.pi, 50) 
contour_levels = np.linspace(-1, 1, 19)
r, theta = np.meshgrid(zeniths, azimuths)

u = 4/15 * (r**2 - 1/r**2) * np.cos(2*theta) 

fig, ax = plt.subplots(subplot_kw = dict(projection = 'polar'))
ax.set_xlim(0, 2*np.pi) # setting up the independent (angle) axis
ax.set_xticks(np.arange(0, 2*np.pi, 2*np.pi/4)) # setting up the ticks displayed
ax.set_ylim(0, 2) # setting up the dependent (r) axis to go from 0 to 2
ax.set_yticks(np.linspace(1, 2, 2)) # setting up the r ticks to be just 1 and 2
CS = ax.contour(theta, r, u, 15, cmap = 'plasma') # 15 contours, 'plasma' colour map
plt.clabel(CS, inline = 1, fontsize = 10) # you can add contour labels
CS2 = ax.contour(theta, r, r, levels = [1.001, 1.999], colors = 'black') # adding heavy black circles at the boundaries
plt.show()

fig, ax = plt.subplots(subplot_kw = dict(projection = 'polar'))
ax.set_xlim(0, 2*np.pi)
ax.set_xticks(np.arange(0, 2*np.pi, 2*np.pi/4))
ax.set_ylim(0, 2)
ax.set_yticks(np.linspace(1, 2, 2))
CS = ax.contourf(theta, r, u, 30, cmap = 'plasma') # 30 contours this time, 'plasma' colour map
CS2 = ax.contour(theta, r, r, levels=[1.001, 1.999], colors = 'black')
cbar=fig.colorbar(CS) # create a colorbar legend
cbar.ax.set_yticks(np.linspace(-1, 1, 11)) # put a reasonable set of tics on the legend
plt.show()
```
